```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
football = pd.read_csv("C:/Users/User/Downloads/dataset - 2020-09-24.csv")
```

```python
print(football.dtypes)
```

```python
sns.lineplot(x='Yellow cards',y='Red cards',data = football)
sns.set_style("dark")
plt.title('Graph illustrating Yellow cards vs Red cards')
plt.show()
```

```python
sns.lineplot(x='Passes',y='Assists',data = football)
sns.set_style("darkgrid")
plt.title("A graph showing preview of Appearances vs Assits obtained")
plt.show()
```

```python
sns.lineplot(x ='Appearances',y='Fouls',data=football)
sns.set_style("ticks")
plt.title("A graph for Appearances vs Fouls")
sns.set_context("talk")
plt.show()
```

```python
sns.distplot(football['Shots'])
sns.set_style("ticks")
plt.title("A distribution graph for Shots")
sns.set_context("talk")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.set_context("talk")
sns.distplot(football['Tackles'])
plt.title('Distribution graph for Tackles')
plt.show()
```

```python
sns.kdeplot(data=football['Clean sheets'],shade=True)
plt.title('A kde graph for Clean sheets')
plt.show()
```

```python
sns.kdeplot(data=football['Big chances missed'],shade=True)
plt.title('A kde graph for Big chances missed')
plt.show()
```

```python
sns.set_style("ticks")
sns.rugplot(football['Goals conceded'])
sns.set_context("talk")
plt.title('Rugplot  graph for Goals Conceded')
plt.show()
```

```python
sns.set_style("ticks")
sns.rugplot(football['Clearances'])
sns.set_context("talk")
plt.title('Rugplot  graph for Clearances')
plt.show()
```

```python
sns.set_style("ticks")
sns.boxplot(x='Position',y='Age',data=football)
plt.title('Box plot graph showing Relationship between Age vs Club')
plt.show()
```

```python
sns.set_style("ticks")
sns.boxplot(x='Position',y='Age',data=football)
plt.title('Box plot graph showing Relationship between Age vs Club')
plt.show()
```

```python
sns.set_style("ticks")
sns.violinplot(x='Position', y='Red cards', data=football)
plt.title('violin_graph showing Relationship between Position vs Red cards')
plt.show()
```

```python
sns.set_style("ticks")
sns.violinplot(x='Position', y='Assists', data=football)
plt.title('violin_graph showing Relationship between Position vs Assists')
plt.show()
```

```python
sns.set_style("ticks")
sns.swarmplot(x='Position', y='Wins', data=football,size = 5)
plt.title('Swarm_graph showing Relationship between Position vs Wins ')
plt.show()
```

```python
sns.set_style("ticks")
sns.swarmplot(x='Position', y='Fouls', data=football)
plt.title('Swarm_graph showing Relationship between Position vs Fouls ')
plt.show()
```

```python
sns.set_style("darkgrid")
sns.jointplot(x='Red cards', y='Yellow cards', data=football, 
kind='scatter')
plt.title('Joint_graph _graph for Reds cards vs Yellow cards')
plt.show()
```

```python
sns.set_style("darkgrid")
sns.jointplot(x='Passes', y='Wins', data=football, 
kind='scatter')
plt.title('Joint_graph _graph for Passes vs Wins')
plt.show()
```

```python
sns.pairplot(football.head(3),hue = 'Position')
plt.show()
```

```python
sns.pairplot(football.head(3),hue = 'Nationality')
plt.show()
```
